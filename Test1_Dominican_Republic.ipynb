```python
import urllib.request
import json
import pandas as pd
from datetime import datetime
from dateutil.relativedelta import relativedelta

def fetch_data():
    try:
        # Get the current month and year
        current_date = datetime.now()
        three_months_ago = current_date - relativedelta(months=3)
        three_months_ago_str = three_months_ago.strftime("%Y-%m")

        # Construct the URL with current month and year
        url = f"https://apis.sb.gob.do/estadisticas/v2/captaciones/localidad?periodoInicial={three_months_ago_str}&periodoFinal={three_months_ago_str}&entidad=ADEMI&paginas=1&registros=40"

        hdr = {
            'Cache-Control': 'no-cache',
            'Ocp-Apim-Subscription-Key': 'ea733680c6424c869566bb75380167b4',
        }

        req = urllib.request.Request(url, headers=hdr)
        req.get_method = lambda: 'GET'
        response = urllib.request.urlopen(req)
        data = response.read()

        # Convert the JSON response to a Python dictionary
        data_dict = json.loads(data)

        # Normalize the data using pandas
        df = pd.json_normalize(data_dict)
        return df

    except Exception as e:
        print("Error fetching data:", e)
        return None

def main():
    # Fetch the data
    df = fetch_data()

    if df is not None:
        # Do something with the data, like save it to a file or store it in a database
        print("Data fetched successfully:")
        print(df)
    else:
        print("Failed to fetch data.")

if __name__ == "__main__":
    main()
```
